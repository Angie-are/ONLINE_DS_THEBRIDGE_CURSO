```python
# Importamos los paquetes y mdulos necesarios
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## #1

### #1.1

Carga el dataset de casas de California que est en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripcin. Deshazte de las filas con nulos.

```python
# Cargar el dataset de casas de California
file_path = "./data/california_cities.csv"  # Asegrate de que la ruta es correcta
df_california = pd.read_csv(file_path)

# Muestra las primeras filas del dataset
print(df_california.head())

# Describe el dataset
print(df_california.describe())

# Eliminar filas con valores nulos
df_california_cleaned = df_california.dropna()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Verificar las columnas disponibles en el DataFrame
print(df_california_cleaned.columns)
```

```python
import matplotlib.pyplot as plt

# Graficamos las ciudades ms pobladas
plt.figure(figsize=(10,6))

# Usamos 'city' y 'population_total' para las etiquetas y valores
plt.bar(df_california_cleaned['city'], df_california_cleaned['population_total'], 
        color=['blue', 'grey', 'olive', 'brown'], alpha=0.5)

# Aadir texto con la poblacin de cada ciudad
for i, row in df_california_cleaned.iterrows():
    plt.text(row['city'], row['population_total'], str(row['population_total']), ha='center', fontsize=10)

# Aadir ttulo y etiquetas
plt.title('Ciudades ms pobladas de California', fontsize=14)
plt.xlabel('Ciudad', fontsize=12)
plt.ylabel('Poblacin', fontsize=12)

# Rotar las etiquetas del eje X para mejorar la visibilidad
plt.xticks(rotation=90)

# Mostrar el grfico
plt.show()

```

```python
# Histograma del rea total ocupada
plt.figure(figsize=(10,6))

# Usamos 'area_total_km2' para el rea total en kilmetros cuadrados
plt.hist(df_california_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black')

# Aadir ttulo y etiquetas
plt.title('Distribucin del rea total ocupada por las ciudades', fontsize=14)
plt.xlabel('rea total ocupada (km)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostrar el grfico
plt.show()

```

```python
# Scatter plot del rea de agua vs poblacin
plt.figure(figsize=(10,6))

# Usamos 'area_water_km2' para el rea de agua y 'population_total' para la poblacin
plt.scatter(df_california_cleaned['area_water_km2'], df_california_cleaned['population_total'], color='green', alpha=0.5)

# Aadir ttulo y etiquetas
plt.title('rea de agua vs Poblacin', fontsize=14)
plt.xlabel('rea de agua (km)', fontsize=12)
plt.ylabel('Poblacin', fontsize=12)

# Mostrar el grfico
plt.show()

```

```python
# Boxplot de la altura de las ciudades
plt.figure(figsize=(10,6))

# Usamos 'elevation_m' para la altura en metros
sns.boxplot(data=df_california_cleaned['elevation_m'], color='orange')

# Aadir ttulo
plt.title('Distribucin de las alturas de las ciudades', fontsize=14)
plt.ylabel('Altura (m)', fontsize=12)

# Mostrar el grfico
plt.show()

```

```python
# Crear una figura con mltiples subgrficas
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Graficar las ciudades ms pobladas
axs[0, 0].bar(df_california_cleaned['city'], df_california_cleaned['population_total'], color=['blue', 'grey', 'olive', 'brown'], alpha=0.5)
for i, row in df_california_cleaned.iterrows():
    axs[0, 0].text(row['city'], row['population_total'], str(row['population_total']), ha='center', fontsize=10)
axs[0, 0].set_title('Ciudades ms pobladas')
axs[0, 0].set_xlabel('Ciudad')
axs[0, 0].set_ylabel('Poblacin')
axs[0, 0].tick_params(axis='x', rotation=90)

# Graficar el histograma del rea total ocupada
axs[0, 1].hist(df_california_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black')
axs[0, 1].set_title('Distribucin del rea total ocupada')
axs[0, 1].set_xlabel('rea total ocupada (km)')
axs[0, 1].set_ylabel('Frecuencia')

# Graficar el scatterplot del rea de agua vs poblacin
axs[1, 0].scatter(df_california_cleaned['area_water_km2'], df_california_cleaned['population_total'], color='green', alpha=0.5)
axs[1, 0].set_title('rea de agua vs Poblacin')
axs[1, 0].set_xlabel('rea de agua (km)')
axs[1, 0].set_ylabel('Poblacin')

# Graficar el boxplot de la altura
sns.boxplot(data=df_california_cleaned['elevation_m'], color='orange', ax=axs[1, 1])
axs[1, 1].set_title('Distribucin de las alturas')

# Ajuste para mejorar la visualizacin
plt.tight_layout()

# Mostrar la figura
plt.show()


```
