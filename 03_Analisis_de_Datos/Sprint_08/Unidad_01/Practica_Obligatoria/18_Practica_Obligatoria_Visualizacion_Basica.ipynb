```python
# Importamos los paquetes y módulos necesarios
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## #1

### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
# Cargar el dataset de casas de California
file_path = "./data/california_cities.csv"  # Asegúrate de que la ruta es correcta
df_california = pd.read_csv(file_path)

# Muestra las primeras filas del dataset
print(df_california.head())

# Describe el dataset
print(df_california.describe())

# Eliminar filas con valores nulos
df_california_cleaned = df_california.dropna()
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Verificar las columnas disponibles en el DataFrame
print(df_california_cleaned.columns)
```

```python
import matplotlib.pyplot as plt

# Graficamos las ciudades más pobladas
plt.figure(figsize=(10,6))

# Usamos 'city' y 'population_total' para las etiquetas y valores
plt.bar(df_california_cleaned['city'], df_california_cleaned['population_total'], 
        color=['blue', 'grey', 'olive', 'brown'], alpha=0.5)

# Añadir texto con la población de cada ciudad
for i, row in df_california_cleaned.iterrows():
    plt.text(row['city'], row['population_total'], str(row['population_total']), ha='center', fontsize=10)

# Añadir título y etiquetas
plt.title('Ciudades más pobladas de California', fontsize=14)
plt.xlabel('Ciudad', fontsize=12)
plt.ylabel('Población', fontsize=12)

# Rotar las etiquetas del eje X para mejorar la visibilidad
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()

```

```python
# Histograma del área total ocupada
plt.figure(figsize=(10,6))

# Usamos 'area_total_km2' para el área total en kilómetros cuadrados
plt.hist(df_california_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black')

# Añadir título y etiquetas
plt.title('Distribución del área total ocupada por las ciudades', fontsize=14)
plt.xlabel('Área total ocupada (km²)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostrar el gráfico
plt.show()

```

```python
# Scatter plot del área de agua vs población
plt.figure(figsize=(10,6))

# Usamos 'area_water_km2' para el área de agua y 'population_total' para la población
plt.scatter(df_california_cleaned['area_water_km2'], df_california_cleaned['population_total'], color='green', alpha=0.5)

# Añadir título y etiquetas
plt.title('Área de agua vs Población', fontsize=14)
plt.xlabel('Área de agua (km²)', fontsize=12)
plt.ylabel('Población', fontsize=12)

# Mostrar el gráfico
plt.show()

```

```python
# Boxplot de la altura de las ciudades
plt.figure(figsize=(10,6))

# Usamos 'elevation_m' para la altura en metros
sns.boxplot(data=df_california_cleaned['elevation_m'], color='orange')

# Añadir título
plt.title('Distribución de las alturas de las ciudades', fontsize=14)
plt.ylabel('Altura (m)', fontsize=12)

# Mostrar el gráfico
plt.show()

```

```python
# Crear una figura con múltiples subgráficas
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Graficar las ciudades más pobladas
axs[0, 0].bar(df_california_cleaned['city'], df_california_cleaned['population_total'], color=['blue', 'grey', 'olive', 'brown'], alpha=0.5)
for i, row in df_california_cleaned.iterrows():
    axs[0, 0].text(row['city'], row['population_total'], str(row['population_total']), ha='center', fontsize=10)
axs[0, 0].set_title('Ciudades más pobladas')
axs[0, 0].set_xlabel('Ciudad')
axs[0, 0].set_ylabel('Población')
axs[0, 0].tick_params(axis='x', rotation=90)

# Graficar el histograma del área total ocupada
axs[0, 1].hist(df_california_cleaned['area_total_km2'], bins=50, color='steelblue', edgecolor='black')
axs[0, 1].set_title('Distribución del área total ocupada')
axs[0, 1].set_xlabel('Área total ocupada (km²)')
axs[0, 1].set_ylabel('Frecuencia')

# Graficar el scatterplot del área de agua vs población
axs[1, 0].scatter(df_california_cleaned['area_water_km2'], df_california_cleaned['population_total'], color='green', alpha=0.5)
axs[1, 0].set_title('Área de agua vs Población')
axs[1, 0].set_xlabel('Área de agua (km²)')
axs[1, 0].set_ylabel('Población')

# Graficar el boxplot de la altura
sns.boxplot(data=df_california_cleaned['elevation_m'], color='orange', ax=axs[1, 1])
axs[1, 1].set_title('Distribución de las alturas')

# Ajuste para mejorar la visualización
plt.tight_layout()

# Mostrar la figura
plt.show()


```
